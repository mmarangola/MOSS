### Testing for Pulse Width Modulation Conversions with Duty Cycle, Frequency, Pulse Length, etc.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
freq = 1000 # pulses per second
duty_cycle = 1 # percent of pulse that logic level is high
duty_Cycle_resolution = 100

# we want to plot the logic level over the total time
t_tot = 1 #seconds
npulses = freq*t_tot

print(f"Total number of pulses: {npulses}")


time = np.arange(0, npulses*duty_Cycle_resolution, 1)
logicVals = np.zeros(npulses*duty_Cycle_resolution) # have res. values for each pulse, because that is our duty cycle constraint

count = 0
for i in range(len(logicVals)//duty_Cycle_resolution):
    for i in range(duty_Cycle_resolution):
        if i < duty_cycle:
            logicVals[count] = 1
        else:
            logicVals[count] = 0
        count = count + 1

        



plt.plot(time, logicVals)
plt.show()

    




```

```python
# working code with duty cycle and frequency!

freq = 5
pulseLen = 10
RES = 100
dc= 2

index = np.arange(0,freq*pulseLen*RES)
y = np.zeros(len(index))

count = 0

for i in range(len(index)//100):
    for j in range(100):
        if j<dc:
            y[count] = 1
        else:
            y[count] = 0
        count = count + 1

plt.plot(index/(freq*RES),y)

plt.show()
```

```python
# smaller timescale

# longer pulse -> increase duty cycle or lower frequency

# variables
freq = 10_000 # maximum potential pulses per second
numPulses = 1 # number of actual pulses
dc= 50 # duty cycle - percent of each cycle that the laser is on

RES = 100
PULSES_LEN = numPulses/freq


index = np.arange(0,freq*PULSES_LEN*RES)
y = np.zeros(len(index))

count = 0

for i in range(len(index)//100):
    for j in range(100):
        if j<dc:
            y[count] = 5
        else:
            y[count] = 0
        count = count + 1

plt.plot(index/(freq*RES),y)
plt.title("Expected Output")
plt.ylabel("V Out (Volts)")
plt.xlabel("time (seconds)")

plt.show()
```
